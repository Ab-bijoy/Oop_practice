Write a Python class named BankAccount.

2.Create a method called deposit that accepts an amount and adds it to the balance.

3.Create a method called withdraw that accepts an amount and subtracts it from the balance.

4.Outside the class, create two separate objects:

    account1 for "Alice" starting with 100.

    account2 for "Bob" starting with 0 (use the default).

5.Deposit 50 into Alice's account.

7.Print both of their final balances.

```python
class BankAccount:
    def __init__(self,acc_holder,balance=0):
        self.acc_holder=acc_holder
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        return f"The current balance is {self.balance}"
    
    def withdraw(self,amount):
        self.balance-=amount
        return f"The current balance is {self.balance}"

alice = BankAccount("Alice",100)
bob = BankAccount("Bob",0)

```

Write a Python program with two classes: Item and ShoppingCart.

2.The ShoppingCart class __init__ should create an attribute called items and set it to an empty list [].

3.Inside ShoppingCart, create a method called add_item that accepts an Item object and appends it to the items list.

4.Inside ShoppingCart, create a method called get_total. This method should loop through the items list, add up all the prices, and return the total sum.

5.Outside the classes, create a ShoppingCart object.

6.Create two Item objects: "Apple" for 1.50 and "Banana" for 2.00.

7.Add both items to the cart using the add_item method.

8.Call get_total and print the final cost.

```python
class Item:
    def __init__(self,name,price):
        self.name = name 
        self.price = price

class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_item(self,item):
        self.items.append(item)
    def get_total(self):
        sum = 0
        for i in self.items:
            sum+=i.price
        return sum

shopcart = ShoppingCart()
```

```python
apple = Item("Apple",1.50)
banana = Item("Banana",2.00)
```

```python
shopcart.add_item(apple)
shopcart.add_item(banana)
```

```python
Amount = shopcart.get_total()
print(Amount)
```

# Exercise 7: Classes and Objects (Internal Math)

Write a Python program using a class named Rectangle.

1.The __init__ method should accept and set width (number) and height (number).

2.Create a method called calculate_area that calculates and returns the area (width multiplied by height).

3.Create a method called calculate_perimeter that calculates and returns the perimeter (width plus height, multiplied by 2).

4.Outside the class, create a Rectangle object with a width of 5 and a height of 10.

5.Call both methods and print their returned results.

```python
class Rectangle:
    def __init__(self,width,height):
        self.width=width
        self.height=height
    
    def calculate_area(self):
        return self.width*self.height
    
    def calculate_perimeter(self):
        return 2*(self.width+self.height)

rectangle = Rectangle(5,10)

print(f"Area: {rectangle.calculate_area()}")
```

```python
print(f"Perimeter: {rectangle.calculate_perimeter()}")
```

6.Outside the class, create a CoffeeMachine object starting with 15 water.

7.Call make_coffee once and print the result.

8.Call make_coffee a second time and print the result.

```python
class CoffeeMachine:

    def __init__(self, water_level):
        self.water_level = water_level
    
    def check_water(self):
        if self.water_level < 10:
            return "Not enough water"
        else:
            return "Water level ok"
    
    def make_coffee(self):
        if self.check_water() == "Water level ok":
            self.water_level-=10
            return "Coffee is ready"
        else:
            return "Please add water"

coffeemachine = CoffeeMachine(15)
```

```python
coffeemachine.make_coffee()
```

Write a Python program using two classes: Task and TodoList.

1.The Task class __init__ should accept a title (string). It should also set an attribute called is_completed to False by default.

2.Inside Task, create a method called complete_task that changes is_completed to True.

3.The TodoList class __init__ should start with an empty list called tasks ([]).

4.Inside TodoList, create a method add_task that accepts a Task object and adds it to the list.

5.Inside TodoList, create a method called get_pending_tasks. This method should loop through the tasks list, find only the tasks where is_completed is False, and return a count (an integer) of how many pending tasks are left.

6.Outside the classes, create a TodoList.

7.Create three Task objects: "Wash dishes", "Do laundry", and "Buy groceries". Add them all to the list.

8.Call complete_task() on the "Wash dishes" object.

9.Print the result of get_pending_tasks() from your to-do list (it should output 2).

```python
class Task:

    def __init__(self,title):
        self.title = title
        self.is_completed = False

    def complete_task(self):
        self.is_completed = True

class TodoList:

    def __init__(self):
        self.tasks = []
    
    def add_task(self,task):
        self.tasks.append(task)
    
    def get_pending_tasks(self):
        counter = 0
        for task in self.tasks:
            if(task.is_completed == False):
                counter+=1
        return counter

Todo = TodoList()

Task1 = Task("Wash dishes")


Task2 = Task("Do laundry")


Task3 = Task("Buy groceries")

Todo.add_task(Task1)
Todo.add_task(Task2)
Todo.add_task(Task3)



```

```python
Task1.complete_task()
```

```python
print(f"Pending task {Todo.get_pending_tasks()}")
```

3.Inside add_box, calculate the total number of items by adding self.item_count and other_box.item_count.

4.Instead of returning a number, create a brand new Box object using that total number, and return that new Box.

7.Call add_box on the first box, passing the second box as the parameter. Save the returned result into a new variable called mega_box.

```python
class Box:
    def __init__(self,item_count=0):
        self.item_count=item_count
    
    def add_box(self,another_box):
        total = self.item_count+another_box.item_count
        new_box=Box()
        new_box.item_count = total
        return new_box

box1 = Box(5)

box2 = Box(10)

newbox = box1.add_box(box2)

print(newbox.item_count)

```

# Exercise 15: Classes and Objects (The Capstone)

Write a Python program using two classes: Student and Classroom.

1.The Student class __init__ should accept and set name (string) and grade (number).

2.The Classroom class __init__ should start with an empty list called students ([]).

3.Inside Classroom, create a method enroll. It accepts a Student object and adds it to the students list.

4.Inside Classroom, create a method get_average_grade. This method should loop through the students list, add up all their grades, and return the mathematical average (total sum divided by the number of students).

    (Hint: you can use len(self.students) to find out how many students are in the list. Assume for this exercise that the list will never be empty).

5.Outside the classes, create a Classroom object.

6.Create three Student objects: "Alice" (90), "Bob" (80), and "Charlie" (85).

7.Enroll all three students into the classroom.

8.Call get_average_grade and print the result.

```python
class Student:
    def __init__(self,name,grade):
        self.name = name
        self.grade = grade
    
```

```python
class Classroom:
    def __init__(self):
        self.students=[]

    def enroll(self,student):
        self.students.append(student)
    
    def get_average_grade(self):
        grade = 0
        for student in self.students:
            grade += student.grade
        average_grade = (grade)/len(self.students)
        return average_grade
```

```python
classroom = Classroom()

student1 = Student("Alice",90)
student2 = Student("Bob",80)
student3 = Student("Charile",85)
```

```python
classroom.enroll(student1)
classroom.enroll(student2)
classroom.enroll(student3)
```

```python
classroom.get_average_grade()
```
