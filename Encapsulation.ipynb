# Exercise 18: Encapsulation (The Python @property Decorator)

Write a Python program using a class named SmartLight.

1.The __init__ method should create a private attribute called __brightness and set it to 50.

2.Above your getter method, write @property. Then, name the method simply brightness(self). Inside, return self.__brightness.

3.Above your setter method, write @brightness.setter. Name the method exactly the same: brightness(self, level).

4.Inside this setter, check if level is between 0 and 100 (inclusive). If it is, update self.__brightness = level. If it isn't, print "Error: Brightness must be 0-100". (Do not use return here, just print the error).

5.Outside the class, create a SmartLight object.

6.Try to set the brightness to 150 just by using an equals sign: my_light.brightness = 150.

7.Set the brightness to 75 using the equals sign: my_light.brightness = 75.

8.Finally, print the brightness just by calling the name: print(my_light.brightness).

```python
class SmartLight:

    def __init__(self):
        self.__brightness = 50
    
    @property
    def brightness(self):
        return self.__brightness
    
    @brightness.setter
    def brightness(self,level):
        if 0 <= level <= 100 :
            self.__brightness = level 
        else:
            print("Error: Brightness must be 0-100")
    
smartlight = SmartLight()

smartlight.brightness = 150
    
smartlight.brightness = 75

    

print(smartlight.brightness)


    

    
```

Write a Python program using a class named GameCharacter.

1.The __init__ method should accept a name (string) and set it as a public attribute (self.name).

2.Also in __init__, create a private attribute __health and set it to 100.

3.Create a public method called take_damage that accepts an amount (number).

4.Inside take_damage, subtract the amount from self.__health.

5.Add a safety check inside the method: if self.__health drops below 0, set it exactly to 0 (so a character never has negative health).

6.Create a read-only getter method (@property) called health that returns self.__health. Do not create a setter. (We only want health to decrease via the take_damage method, never by someone cheating and directly typing character.health = 1000).

7.Outside the class, create a character named "Hero".

8.Make the character take 150 damage using the take_damage method.

9.Print the character's health to prove it stopped exactly at 0.

```python
class GameCharacter:

    def __init__(self,name):
        self.name = name
        self.__health = 100
    
    def take_damage(self,number):
        self.__health= self.__health-number
        if(self.__health <= 0):
            self.__health = 0
    
    @property
    def health(self):
        return self.__health

hero = GameCharacter("Hero")

hero.take_damage(150)
    
    



    
```

# Exercise 21: Encapsulation (Private Methods)

Write a Python program using a class named CreditCard.

1.The __init__ method should accept and set a public balance (number).

2.Create a private method called __check_fraud. It should accept an amount (number). If the amount is strictly greater than 1000, return True (meaning it is fraud). Otherwise, return False.

3.Create a public method called make_purchase that accepts an amount.

4.Inside make_purchase, first call your private method to test the amount: fraud_detected = self.__check_fraud(amount).

5.If the fraud check came back True, do not deduct money. Just return the string "Transaction blocked: Fraud alert!".

6.If the fraud check came back False, subtract the amount from self.balance and return "Transaction successful.".

7.Outside the class, create a CreditCard object with a starting balance of 5000.

8.Call make_purchase for 1500 and print the result.

9.Call make_purchase for 200 and print the result.

```python
class CreditCard:
    def __init__(self,balance):
        self.balance = balance
    
    def __check_fraud(self,number):
        if(number>1000):
            return True
        else:
            return False
    
    def make_purchase(self,amount):
        if self.__check_fraud(amount) == True:
            return "Transaction Blocked:Fraud alert!"
        
        else:
            self.balance-=amount
            return "Transaction successful"

obj = CreditCard(5000)
```

```python
obj.make_purchase(1500)
```

```python
obj.make_purchase(200)
```

4.If both conditions are true, update self.__email and self.__password with the new data, and return the string "Registration successful".

Write a Python program using a class named SecretDocument.

1.The __init__ method should accept and set a content (string) into a private attribute called __content.

2.Create a standard @property called content that returns self.__content.

3.Do not create a setter.

4.Create a deleter decorator by writing @content.deleter above a method named content(self).

5.Inside the deleter method, do not actually delete the variable. Instead, overwrite self.__content with the string "REDACTED".

6.Outside the class, create a SecretDocument with the text "Launch codes are 0000".

7.Print the content of the document.

8.Use Python's built-in del keyword to trigger your hidden deleter method: del my_document.content.

9.Print the content of the document again to prove it was successfully redacted.

```python
class SecretDocument:
    def __init__(self,content):
        self.__content = content 
    
    @property
    def content(self):
        return self.__content

    @content.deleter 
    def content(self):
        self.__content = "REDACTED"

obj = SecretDocument("Launch codes are 0000")
```

```python
print(obj.content)
```

```python
del obj.content
```

```python
print(obj.content)
```

# Exercise 24: Encapsulation (Dynamic Properties)

Write a Python program using a class named ShoppingCart.

1.The __init__ method should create a private list called __prices and set it to [].

2.Create a public method called add_item that accepts a price (number). Inside, append the price to self.__prices.

3.Create a read-only @property called total_price.

4.Inside the total_price method, do not return a single variable. Instead, dynamically calculate and return the total sum of everything inside the self.__prices list. (Hint: you can use Python's built-in sum() function here, e.g., sum(self.__prices)).

5.Do not create a setter or deleter.

6.Outside the class, create a ShoppingCart object.

7.Use add_item to add a 10 dollar item.

8.Use add_item to add a 25 dollar item.

9.Print the total_price of the cart.

```python
class ShoppingCart:
    def __init__(self):
        self.__prices = []

    def add_item(self,amount):
        self.__prices.append(amount)
    
    @property
    def total_price(self):
        sum = 0
        for price in self.__prices:
            sum+=price
        return sum
    
    
    
obj = ShoppingCart()
```

```python
obj.add_item(10)
obj.add_item(25)

print(obj.total_price)

```

# Exercise 25: Encapsulation (The Capstone)

Write a Python program using a class named SecureBank.

1.The __init__ method should create a private attribute __balance and set it to 0.

2.Create a read-only @property called balance that returns the __balance.

3.Create a public method called deposit that accepts an amount.

    Validation check: If amount is strictly greater than 0, add it to __balance and return "Deposit successful".

    Otherwise, return "Invalid deposit amount".

4.Create a public method called withdraw that accepts an amount.

    Validation check: First, check if amount is greater than 0. Second, check if amount is less than or equal to self.__balance.

    If both are true (Positive Logic!), subtract the amount and return "Withdrawal successful".

    Otherwise, return "Invalid withdrawal or insufficient funds".

5.Outside the class, create a SecureBank object.

6.Try to deposit -50 and print the result.

7.Deposit 100 and print the result.

8.Try to withdraw 200 and print the result.

9.Withdraw 40 and print the result.

10.Finally, print the read-only balance property to prove it correctly sits at 60.

```python
class SecureBank:
    def __init__(self):
        self.__balance = 0
    
    @property
    def balance(self):
        return self.__balance
    
    def deposit(self,amount):
        if(amount>0):
            self.__balance+=amount
            print("Deposit Successful")
        else:
            print("Invalid deposit amount")
    
    def withdraw(self,amount):
        if(amount>0 and amount<=self.__balance):
            self.__balance-=amount
            print("Withdrawal successful")
        else:
            print("Invalid wihtdrawal or insufficient funds")

obj = SecureBank()
        
```

```python
obj.withdraw(200)
```

```python
obj.withdraw(40)
```
